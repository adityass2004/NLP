## Regular Expression

### 1. Email & Indian Phone Number Extraction

```python
text = """
Contact us at support@example.com or admin.test@domain.co.in.
Call +91 9709303105 or 9876543210 for help.
"""

emails = re.findall(r'[a-zA-Z0-9._%+-]+@[a-zA-Z.-]+\.[a-zA-Z]{2,}', text)
phones = re.findall(r'(?:\+91\s?|91\s?)?[6-9]\d{9}', text)

print("Emails:", emails)
print("Phone Numbers:", phones)
```

### 2. Text Cleaning

```python
text = "Visit https://example.com!!! NLP   is   awesome."

text = re.sub(r'https?://\S+', '', text)
text = re.sub(r'[^a-zA-Z\s]', '', text)
text = re.sub(r'\s+', ' ', text)

print(text.strip())
```

### 3. Password Validation

```python
password = "Strong@123"

pattern = r'^(?=.*[A-Z])(?=.*\d)(?=.*[@$!%*?&])[A-Za-z\d@$!%*?&]{8,}$'

if re.match(pattern, password):
    print("Valid Password")
else:
    print("Invalid Password")
```
