```python
import re
import nltk
import math
from collections import Counter
from sklearn.feature_extraction.text import TfidfVectorizer

nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
```

## TF-IDF & Word-Level Analysis

### 1. EManual TF Calculation

```python
doc = "I love NLP and I love machine learning".lower().split()
count = Counter(doc)

tf_love = count['love'] / len(doc)
tf_nlp = count['nlp'] / len(doc)

print("TF(love):", tf_love)
print("TF(nlp):", tf_nlp)
```

### 2. TF-IDF Calculation

```python
documents = [
    "NLP is fun",
    "NLP is powerful",
    "I love NLP"
]

vectorizer = TfidfVectorizer()
tfidf = vectorizer.fit_transform(documents)

print(vectorizer.get_feature_names_out())
print(tfidf.toarray())
```

### 3. Keyword Extraction using TF-IDF

```python
reviews = [
    "This phone has excellent battery life",
    "Battery performance of this phone is great",
    "Excellent camera and battery"
]

vectorizer = TfidfVectorizer(stop_words='english')
tfidf = vectorizer.fit_transform(reviews)

feature_names = vectorizer.get_feature_names_out()

for i, review in enumerate(tfidf.toarray()):
    top_words = sorted(zip(feature_names, review), key=lambda x: x[1], reverse=True)[:5]
    print(f"Review {i+1} Keywords:", top_words)
```
