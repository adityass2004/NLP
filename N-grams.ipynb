```python
import nltk
import os

nltk_data_dir = os.path.join(os.getcwd(), "nltk_data")
os.makedirs(nltk_data_dir, exist_ok=True)

nltk.data.path.append(nltk_data_dir)


nltk.download("punkt", download_dir=nltk_data_dir)
nltk.download("punkt_tab", download_dir=nltk_data_dir)
nltk.download("averaged_perceptron_tagger", download_dir=nltk_data_dir)
nltk.download("averaged_perceptron_tagger_eng", download_dir=nltk_data_dir)
```

## N-grams

### 1. Bigram Generator

```python
sentence = "NLP is very interesting"
tokens = nltk.word_tokenize(sentence)

bigrams = list(nltk.bigrams(tokens))
print(bigrams)
```

### 2. Trigram Frequency Counter

```python
text = "NLP is fun and NLP is powerful and NLP is interesting"
tokens = nltk.word_tokenize(text)

trigrams = list(nltk.trigrams(tokens))
freq = Counter(trigrams)

print(freq.most_common(3))
```

### 3. POS Tagging

```python
sentence = "The quick brown fox jumps over the lazy dog"
tokens = nltk.word_tokenize(sentence)

pos_tags = nltk.pos_tag(tokens)
for word, tag in pos_tags:
    print(word, "→", tag)
```

### 3. POS-Based Filtering

```python
text = "The programmer writes clean code and fixes bugs"
tokens = nltk.word_tokenize(text)
tags = nltk.pos_tag(tokens)

nouns = [word for word, tag in tags if tag.startswith('NN')]
verbs = [word for word, tag in tags if tag.startswith('VB')]

print("Nouns:", nouns)
print("Verbs:", verbs)
```
